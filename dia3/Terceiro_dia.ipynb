# Analisados Dados de empretismos

```python
import pandas as pd
```

```python
df = pd.read_csv('../Dia_1-Importando_dados/Datasets/tabela_completa14marco.csv')
```

```python
df.shape
```

```python
df.data_emprestimo =  pd.to_datetime(df.data_emprestimo)
```

```python
df['ano_emprestimo'] = df.data_emprestimo.dt.year
```

```python
grupo = df.groupby('ano_emprestimo')['ano_emprestimo']
```

```python
grupo.value_counts()
```

## Grafico mostra que existiu um queda consideravel em 2020

```python
import matplotlib.pyplot as plt
```

```python
contagem = grupo.count()
i = contagem.index
x = contagem.values
```

```python
fig, ax  = plt.subplots(figsize=(6,2))
plt.style.use('_mpl-gallery')
ax.plot(i,x, linewidth=2.0)
ax.set_title('Ano x Emprestimos')
```

## Visualização por mes

```python
df['mes_emprestimo'] = df.data_emprestimo.dt.month_name()
```

```python
grupo_por_mes = df.groupby('mes_emprestimo')['mes_emprestimo']
```

```python
contagem_mes = grupo_por_mes.count()
a = contagem_mes.index.sort
b = contagem_mes.values
contagem_mes.index.sort
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.plot(a, b)
ax.set_title('Emprestimos por mês')
fig.
```

#### Dezembro e janeiro, são os meses que tem menos emprestimo de livros e depois Julho, devido ser os meses de férias

## Verificando os horarios que tem mais emprestimos dos livros

```python
df['hora_emprestimo'] = df.data_emprestimo.dt.hour
```

```python
grupo_hora = df.groupby('hora_emprestimo')['hora_emprestimo']
```

```python
contagem_hora = grupo_hora.count()
x = contagem_hora.index
y = contagem_hora.values
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(x,y,)
```

### Analisando os horarios com menos frequência de emprestimos são:
Das 7 até as 9 da amanha,13h, 19h até as 21h.
