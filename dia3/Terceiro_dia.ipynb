# Analisados Dados de empretismos

```python
import pandas as pd
```

```python
df = pd.read_csv('../Dia_1-Importando_dados/Datasets/tabela_completa14marco.csv')
```

```python
df.shape
```

```python
df.data_emprestimo =  pd.to_datetime(df.data_emprestimo)
```

```python
df['ano_emprestimo'] = df.data_emprestimo.dt.year
```

```python
grupo = df.groupby('ano_emprestimo')['ano_emprestimo']
```

```python
grupo.value_counts()
```

## Grafico mostra que existiu um queda consideravel em 2020

```python
grupo.value_counts().plot(figsize=(15,5))
```

## Visualização por mes

```python
df['mes_emprestimo'] = df.data_emprestimo.dt.month_name()
```

```python
grupo_por_mes = df.groupby('mes_emprestimo')['mes_emprestimo']
```

```python
grupo_por_mes.value_counts()
```

```python
grupo_por_mes.value_counts().plot(kind='line',figsize=(15,5))
```
